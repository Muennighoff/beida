```python
import sqlite3
from tkinter import *

conn = sqlite3.connect("university.db")
c = conn.cursor()

for statement in open("university.sql").read().split("\n\n"):
    c.execute(statement)
    
conn.commit()
conn.close()
```

```python
# Build GUI
def execute_sql(sql_query):
    conn = sqlite3.connect("university.db")
    c = conn.cursor()
    c.execute(sql_query)
    
    records = c.fetchall()
    
    conn.commit()
    conn.close()
    return records
    

class Window:
    def __init__(self, wname="university"):
        # Basic Configuration
        self.root = Tk()
        self.root.geometry("1400x500")
        self.root.configure(background="white")
        self.root.title(wname)

        # Adding student entry to database
        self.sid_label = Label(self.root, text="Student ID")
        self.sid_label.grid(row=0, column=0, padx=20)
        self.studentid = Entry(self.root, font=("Purisa",12))
        self.studentid.grid(row=0, column=1, padx=20, pady=5)
        
        self.g_label = Label(self.root, text="Gender")
        self.g_label.grid(row=1, column=0, padx=20)
        self.gender = Entry(self.root, font=("Purisa",12))
        self.gender.grid(row=1, column=1, padx=20, pady=5)
        
        self.a_label = Label(self.root, text="Age")
        self.a_label.grid(row=2, column=0, padx=20)
        self.age = Entry(self.root, font=("Purisa",12))
        self.age.grid(row=2, column=1, padx=20, pady=5)
        
        self.m_label = Label(self.root, text="Major")
        self.m_label.grid(row=3, column=0, padx=20)
        self.major = Entry(self.root, font=("Purisa",12))
        self.major.grid(row=3, column=1, padx=20, pady=5)

        self.cname_label = Label(self.root, text="Class")
        self.cname_label.grid(row=4, column=0, padx=20)
        self.classname = Entry(self.root, font=("Purisa",12))
        self.classname.grid(row=4, column=1, padx=20, pady=5)
        
        self.insert_btn = Button(self.root, text="Add Student", command=self.insert_std)
        self.insert_btn.grid(row=5, column=0, columnspan=2, padx=20, pady=20)
        
        
        # Adding course entry to database
        self.cid_label = Label(self.root, text="Course ID")
        self.cid_label.grid(row=0, column=2, padx=20)
        self.courseid = Entry(self.root, font=("Purisa",12))
        self.courseid.grid(row=0, column=3, padx=20, pady=5)
        
        self.tid_label = Label(self.root, text="Teacher ID")
        self.tid_label.grid(row=1, column=2, padx=20)
        self.teacherid = Entry(self.root, font=("Purisa",12))
        self.teacherid.grid(row=1, column=3, padx=20, pady=5)
        
        self.d_label = Label(self.root, text="Department")
        self.d_label.grid(row=2, column=2, padx=20)
        self.department = Entry(self.root, font=("Purisa",12))
        self.department.grid(row=2, column=3, padx=20, pady=5)
        
        self.c_insert_btn = Button(self.root, text="Add Course", command=self.insert_crs)
        self.c_insert_btn.grid(row=5, column=2, columnspan=2, padx=20, pady=20)

        # Adding enrolled entry to database
        self.cid2_label = Label(self.root, text="Course ID")
        self.cid2_label.grid(row=0, column=4, padx=20)
        self.courseid2 = Entry(self.root, font=("Purisa",12))
        self.courseid2.grid(row=0, column=5, padx=20, pady=5)
        
        self.sid2_label = Label(self.root, text="Student ID")
        self.sid2_label.grid(row=1, column=4, padx=20)
        self.studentid2 = Entry(self.root, font=("Purisa",12))
        self.studentid2.grid(row=1, column=5, padx=20, pady=5)
        
        self.grade_label = Label(self.root, text="Grade")
        self.grade_label.grid(row=2, column=4, padx=20)
        self.grade = Entry(self.root, font=("Purisa",12))
        self.grade.grid(row=2, column=5, padx=20, pady=5)
        
        self.c_insert_btn = Button(self.root, text="Add Grade", command=self.insert_grd)
        self.c_insert_btn.grid(row=5, column=4, columnspan=2, padx=20, pady=20)

        # Look-Up Student grades
        self.student = Entry(self.root, font=("Purisa",12))
        self.student.grid(row=6, column=1, padx=20, pady=5)
        
        self.s_label = Label(self.root, text="Student ID")
        self.s_label.grid(row=6, column=0, padx=20)

        self.grade_btn = Button(self.root, text="Grade Lookup", command=self.grade_lookup)
        self.grade_btn.grid(row=7, column=0, columnspan=2, padx=20, pady=20)
        
        
        # Global out messages
        self.out_label = Label(self.root, text="Out Messages", width=30)
        self.out_label.grid(row=9, column=0, columnspan=1, rowspan=1, padx=20, pady=20)
        self.out = Label(self.root, text="", width=100)
        self.out.grid(row=9, column=1, columnspan=4, rowspan=1, padx=20, pady=20)
    
    def insert_std(self):
        try:
            sql_query = f"""
            INSERT INTO students
            VALUES
            ({int(self.studentid.get())}, "{str(self.gender.get())}", 
            {int(self.age.get())}, "{str(self.major.get())}", "{str(self.classname.get())}");
            """
            records = execute_sql(sql_query)
            self.out["text"] = records
        except Exception as e:
            self.out["text"] = e

        # Clear box
        self.studentid.delete(0, END)
        self.gender.delete(0, END)
        self.age.delete(0, END)
        self.major.delete(0, END)
        self.classname.delete(0, END)
        
    def insert_crs(self):
        try:
            sql_query = f"""
            INSERT INTO course
            VALUES
            ({int(self.courseid.get())}, {int(self.teacherid.get())}, "{str(self.department.get())}");
            """
            records = execute_sql(sql_query)
            self.out["text"] = records
        except Exception as e:
            self.out["text"] = e

        # Clear box
        self.courseid.delete(0, END)
        self.teacherid.delete(0, END)
        self.department.delete(0, END)
        
    def insert_grd(self):
        try:
            sql_query = f"""
            INSERT INTO enrolled
            VALUES
            ({int(self.studentid2.get())}, {int(self.courseid2.get())}, "{str(self.grade.get())}");
            """
            records = execute_sql(sql_query)
            self.out["text"] = records
        except Exception as e:
            self.out["text"] = e

        # Clear box
        self.studentid2.delete(0, END)
        self.courseid2.delete(0, END)
        self.grade.delete(0, END)
        
    def grade_lookup(self):        
        try:
            sql_query = f"""
            SELECT (SELECT course.department FROM course WHERE course.courseid = enrolled.courseid), 
                    enrolled.grade
            FROM enrolled
            WHERE enrolled.studentid = {self.student.get()}
            """
            records = execute_sql(sql_query)
            
            
            self.out["text"] = records
        except Exception as e:
            self.out["text"] = e

        # Clear box
        self.student.delete(0, END)
```

```python
w = Window("My window")
w.root.mainloop()
```
